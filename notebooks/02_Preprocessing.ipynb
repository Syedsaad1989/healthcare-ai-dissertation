```python
## Purpose of this notebook

This notebook prepares the Cleveland Heart Disease dataset for machine learning and deep learning experiments.

The preprocessing pipeline is designed to ensure that the data are clean, consistent and reproducible before model development.
The same workflow will later be applied to the diabetes and breast cancer datasets to support fair comparison across multiple clinical datasets.
```

```python
import pandas as pd
import numpy as np

from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
PROJECT_ROOT = Path.cwd().parent

DATA_PATH = PROJECT_ROOT / "data"
RAW_PATH = DATA_PATH / "raw"
PROCESSED_PATH = DATA_PATH / "processed"

PROCESSED_PATH.mkdir(exist_ok=True)
# Dataset configuration
DATASET_NAME = "heart_disease"
DATASET_FILE = "processed.cleveland.data"
```

```python
columns = [
    "age",
    "sex",
    "cp",
    "trestbps",
    "chol",
    "fbs",
    "restecg",
    "thalach",
    "exang",
    "oldpeak",
    "slope",
    "ca",
    "thal",
    "target"
]
```

```python
df = pd.read_csv(
    RAW_PATH / DATASET_NAME / DATASET_FILE,
    names=columns,
    na_values="?"
)

df.head()
```

```python
## Dataset Overview

Before preprocessing, the dataset is inspected to understand its size, structure, data types and potential quality issues.
This provides an overview of the data and identifies any missing values or incorrect data types that require further processing.
```

```python
## Target Variable Transformation

The original Cleveland Heart Disease dataset contains five diagnostic categories:

- 0 = No heart disease
- 1–4 = Presence of heart disease

Following common practice in the literature and the approved dissertation proposal, the target variable is converted 
into a binary classification problem.

Binary classification allows consistent comparison between classical machine learning models and deep neural networks 
while reflecting the practical clinical task of identifying whether heart disease is present.
```

```python
## Binary Target Transformation

The original Cleveland dataset contains five diagnostic classes:

0 = No Heart Disease

1–4 = Presence of Heart Disease

Following common practice in the literature, and consistent with the approved dissertation proposal, 
the target variable is converted into a binary classification problem:

- 0 = No Disease
- 1 = Disease

This transformation supports clinically meaningful evaluation using machine learning classifiers.
```

```python
imputer = SimpleImputer(strategy="most_frequent")

df[["ca", "thal"]] = imputer.fit_transform(df[["ca", "thal"]])
```

```python
X = df.drop("target", axis=1)

y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
train_df = pd.DataFrame(X_train_scaled, columns=X.columns)
train_df["target"] = y_train.reset_index(drop=True)

test_df = pd.DataFrame(X_test_scaled, columns=X.columns)
test_df["target"] = y_test.reset_index(drop=True)

train_df.to_csv(
    PROCESSED_PATH / "heart_train_processed.csv",
    index=False
)

test_df.to_csv(
    PROCESSED_PATH / "heart_test_processed.csv",
    index=False
)

print("Processed datasets saved successfully.")
```

```python
print("Training set:", train_df.shape)

print("Testing set:", test_df.shape)

train_df.head()
```

```python
## Missing Value Treatment

The Cleveland Heart Disease dataset contains a small number of missing values in the **ca** and **thal** variables.

Removing these records would unnecessarily reduce the sample size. Therefore, missing values are imputed using
the most frequent category, preserving all observations while maintaining the categorical nature of these variables.
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="most_frequent")

df[["ca", "thal"]] = imputer.fit_transform(df[["ca", "thal"]])
```

```python
df.isnull().sum()
```

```python
X = df.drop("target", axis=1)
y = df["target"]

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)
```

```python
## Train–Test Split

The dataset is divided into training (80%) and testing (20%) subsets.

Stratified sampling is used to preserve the proportion of patients with and without heart disease in both subsets. 
This ensures that model evaluation is representative and reproducible.

A fixed random state of 42 is used to support reproducibility.
```

```python
print("Training target distribution:")
print(y_train.value_counts())

print("\nTesting target distribution:")
print(y_test.value_counts())
```

```python
## Feature Standardisation

Machine learning algorithms such as Logistic Regression and Deep Neural Networks are sensitive to differences in feature scales.
Therefore, numerical features are standardised using the StandardScaler.

The scaler is fitted only on the training data to prevent information leakage from the test set.
The same fitted scaler is then applied to the testing data.
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
import joblib

MODEL_PATH = PROJECT_ROOT / "models"

MODEL_PATH.mkdir(exist_ok=True)

joblib.dump(
    scaler,
    MODEL_PATH / "heart_scaler.pkl"
)

print("Scaler saved successfully.")
```

```python
X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X.columns,
    index=X_train.index
)

X_test_scaled = pd.DataFrame(
    X_test_scaled,
    columns=X.columns,
    index=X_test.index
)
```

```python
X_train_scaled.head()
```

```python
train_processed = X_train_scaled.copy()
train_processed["target"] = y_train

test_processed = X_test_scaled.copy()
test_processed["target"] = y_test
```

```python
train_processed.to_csv(
    PROCESSED_PATH / "heart_train_processed.csv",
    index=False
)

test_processed.to_csv(
    PROCESSED_PATH / "heart_test_processed.csv",
    index=False
)

print("Processed datasets saved successfully!")
```
