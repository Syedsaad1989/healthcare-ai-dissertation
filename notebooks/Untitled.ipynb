```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
PROJECT_ROOT = Path("..")

DATA_PATH = PROJECT_ROOT / "data" / "raw" / "heart_disease"

print(DATA_PATH)
```

```python
# -----------------------------
# Load Heart Disease Dataset
# -----------------------------

columns = [
    "age",
    "sex",
    "cp",
    "trestbps",
    "chol",
    "fbs",
    "restecg",
    "thalach",
    "exang",
    "oldpeak",
    "slope",
    "ca",
    "thal",
    "target"
]

heart_df = pd.read_csv(
    DATA_PATH / "processed.cleveland.data",
    header=None,
    names=columns,
    na_values="?"
)

heart_df.head()
```

```python
import matplotlib.pyplot as plt

heart_df.hist(
    figsize=(18,12),
    bins=20
)

plt.suptitle(
    "Distribution of Numerical Features in the Heart Disease Dataset",
    fontsize=18
)

plt.tight_layout()

plt.show()
```

```python
### Initial Observations

The distributions indicate that several variables, including age and maximum heart rate, are approximately continuous,
whereas clinical indicators such as chest pain type, fasting blood sugar, and thalassemia are categorical despite being encoded numerically.

Certain variables exhibit skewed distributions, suggesting that preprocessing techniques such as feature scaling and careful 
handling of outliers may be required before model development.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

heart_df["target"].value_counts().sort_index().plot(
    kind="bar"
)

plt.title("Heart Disease Class Distribution")
plt.xlabel("Target Class")
plt.ylabel("Number of Patients")

plt.show()

print(heart_df["target"].value_counts().sort_index())
```

```python
### Interpretation

The original target variable contains five classes representing increasing levels of heart disease severity.
Visual inspection indicates a substantial class imbalance, with Class 0 (no heart disease) accounting for the
largest proportion of observations and Class 4 containing only a small number of cases.

This imbalance would make reliable multi-class classification challenging due to the limited number of observations
in the higher severity categories. Consequently, in accordance with the research objectives of this dissertation,
the target variable will be converted into a binary outcome during preprocessing, where Class 0 represents the absence
of heart disease and Classes 1–4 collectively represent the presence of heart disease.

This binary formulation better reflects the objective of early clinical disease detection and is consistent with common 
practice in the medical machine learning literature.
```

```python
counts = heart_df["target"].value_counts().sort_index()

plt.figure(figsize=(8,5))

bars = plt.bar(
    counts.index.astype(str),
    counts.values
)

plt.title("Distribution of Heart Disease Severity Classes")
plt.xlabel("Heart Disease Class")
plt.ylabel("Number of Patients")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 2,
        str(int(height)),
        ha="center",
        fontsize=10
    )

plt.tight_layout()
plt.show()
```
